```python
import re

def tokenize(text: str) -> list[str]:
    """Converts text to lowercase and splits into words, removing punctuation."""
    text = text.lower()
    # Use regex to find words, splitting on non-alphanumeric characters
    tokens = re.findall(r'\b\w+\b', text)
    return tokens

def word_ngrams(tokens: list[str], n: int) -> list[tuple[str, ...]]:
    """Generates word n-grams from a list of tokens."""
    if n <= 0:
        raise ValueError("n must be a positive integer")
    if len(tokens) < n:
        return []

    ngrams = []
    for i in range(len(tokens) - n + 1):
        ngrams.append(tuple(tokens[i : i + n]))
    return ngrams

def main() -> None:
    text = "the cat sat on the mat and the cat ate fish"
    tokens = tokenize(text)

    print("Text:", text)
    print("Tokens:", tokens)
    print()

    # Demonstrate 2-grams (bigrams)
    bigrams = word_ngrams(tokens, 2)
    print("Bigrams:", bigrams)

    # Demonstrate 3-grams (trigrams)
    trigrams = word_ngrams(tokens, 3)
    print("Trigrams:", trigrams)

# Call the main function to run the demonstration
main()
```
